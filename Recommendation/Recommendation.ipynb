```python
books_df.rename(columns={'Book.Title': 'Book_Title'}, inplace=True)
books_df.rename(columns={'Book.Rating': 'Book_Rating'}, inplace=True)
books_df.rename(columns={'User.ID': 'User_ID'}, inplace=True)
books_df.head()
```

```python
book_user_df = books_df.pivot_table(index='User_ID',columns='Book_Title',values='Book_Rating')
book_user_df
```

```python
book_user_df.fillna(0.000,inplace=True)
book_user_df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine , correlation
```

```python
user_sim = 1 - pairwise_distances(book_user_df.values ,metric='cosine')
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
pd.set_option('display.float_format', '{:.2f}'.format)
user_sim_df
```

```python
user_sim_df.index = list(book_user_df.index)
user_sim_df.columns = list(book_user_df.index)
user_sim_df
```

```python
user_sim_df.sort_values([8],ascending=False).head(100)
```

```python
def give_reco(customer_id):
    tem = list(user_sim_df.sort_values(customer_id, ascending=False).head(100).index)
    book_list = []

    for i in tem:
        book_list = list(books_df[books_df["User_ID"] == i]["Book_Title"])

    return set(book_list) - set(books_df[books_df["User_ID"] == customer_id]["Book_Title"])
```

```python
give_reco(278846)
```
